# Data Preprocessing:

```python
# to drop the duplicated values:

df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

```python
# dropping the columns with too many missing data, which is in this case is above 90%:

df.drop(columns=['veil-type','spore-print-color','veil-color','stem-root'], inplace = True)
```

```python
df.info()
```

```python
# to fill the missing values with the mode in the rest of the columns having null values.

for i in df.columns:
    if df[i].isnull().sum()>0:
        df[i].fillna(df[i].mode()[0], inplace = True)
```

```python
df.isnull().sum()
```

```python
# Encoding the categorical variables into numerical with label encoding.

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in df.columns:
    if df[i].dtype == 'object':
        df[i] = le.fit_transform(df[i])
```

```python
df.head()
```

```python
# splitting the dataset into features(x) and target(y) for model fitting.

from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['class'])
y = df['class']
```

```python
x
```

```python
y
```

```python
# splitting the dataset into training and the testing sets.

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
y_train
```

# Data Modelling:

```python
# model training and evaluation with logistic regression.

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

LR = LogisticRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred = LR.predict(x_test)
```

```python
accLR = accuracy_score(y_test,y_pred)
accLR
```

```python
confLR = confusion_matrix(y_test,y_pred)
confLR
```

```python
# model training and evaluation with DecisionTreeclassifier.

from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
```

```python
DT.fit(x_train,y_train)
```

```python
y_pred = DT.predict(x_test)
```

```python
accDT = accuracy_score(y_test,y_pred)
accDT
```

```python
confDT = confusion_matrix(y_test,y_pred)
confDT
```

```python
# model training and evaluation with RandomForestClassifier.

from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
```

```python
RF.fit(x_train,y_train)
```

```python
y_pred = RF.predict(x_test)
```

```python
accRF = accuracy_score(y_test,y_pred)
accRF
```

```python
confRF = confusion_matrix(y_test,y_pred)
confRF
```

```python
# model training and evaluation with KNeighborsClassifier.

from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
```

```python
KNN.fit(x_train,y_train)
```

```python
y_pred = KNN.predict(x_test)
```

```python
accKNN = accuracy_score(y_test,y_pred)
accKNN
```

```python
confKNN = confusion_matrix(y_test,y_pred)
confKNN
```
